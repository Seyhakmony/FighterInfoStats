```python
import ast

def count_win_methods(method_list):
    if not isinstance(method_list, list) or len(method_list) == 0:
        return 0, 0, 0
    
    ko_count = 0
    sub_count = 0
    dec_count = 0
    
    for method in method_list:
        method_lower = str(method).lower()
        
        if 'ko' in method_lower or 'tko' in method_lower:
            ko_count += 1
        elif 'submission' in method_lower or 'sub' in method_lower:
            sub_count += 1
        elif 'decision' in method_lower:
            dec_count += 1
    
    return ko_count, sub_count, dec_count

if isinstance(fightersData['method'].iloc[0], str):
    fightersData['method'] = fightersData['method'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) and x != '[]' else [])


fightersData[['ko_t', 'totsub', 'totdec']] = fightersData['method'].apply(
    lambda x: pd.Series(count_win_methods(x))
)
```

```python
fightersData.to_csv('../fightersDatab.csv')
```
