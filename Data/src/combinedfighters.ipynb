```python
fighters = fighters.drop_duplicates(subset='name').reset_index(drop=True)
```

```python
fighters['pfprank'] = fighters['overall'].rank(method='min', ascending=False).astype(int)

pfprank_values = fighters['pfprank']
fighters = fighters.drop('pfprank', axis=1)
fighters.insert(fighters.columns.get_loc('rank') + 1, 'pfprank', pfprank_values)
fighters = fighters.sort_values('pfprank')
```

```python
fighters.to_csv(f"../AllFighters/fighters.csv", index = False)
```
